```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X,y=make_regression(n_samples=10,n_features=1,n_informative=1,n_targets=1,noise=10,random_state=42)

plt.scatter(X,y)
plt.grid(True)
plt.show()
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
m = reg.coef_[0]
```

```python
b = reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
plt.grid(True)
plt.show()
```

```python
# Lets apply Gradient Descent assuming slope is constant m= 78.35
# let's assume the starting value for intercept b = 0
y_pred = ((m*X)+0).reshape(10)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color="green",label='b=0')
plt.legend()
plt.grid(True)
plt.show()
```

```python
m=20.77
b=0

loss_slope1 = -2*np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
# Let take learning rate =0.1
lr=0.1
step_size = loss_slope*lr
step_size
```

```python
# Calculating new intercept(b)

b = b - step_size
b
```

```python
y_pred1 = ((20.77*X)+b).reshape(10)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='black',label='OLS')
plt.plot(X,y_pred1,color='green',label="b={}".format(b))
plt.plot(X,y_pred,color="red",label='b=0')
plt.legend()
plt.grid(True)
plt.show()
```

```python
m = 20.77
b = 0
lr = 0.1

for i in range(100):
    loss_slope = -2 * np.mean(y - m*X.ravel() - b)

    b = b - lr * loss_slope

    print(f"Epoch {i+1}: b = {b:.4f}")
```

```python
y_pred2 = ((20.77*X)+b).reshape(10)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='blue',label='OLS')
plt.plot(X,y_pred2,color='green',label="b={}".format(b))
plt.plot(X,y_pred,color="red",label='b=0')
plt.legend()
plt.grid(True)
plt.show()
```

```python
b = -100
m = 20.77
lr =0.1
epochs=10

for i in range(epochs):
  loss_slope = -2*np.sum(y-m*X.ravel() -b)
  b= b -(lr*loss_slope)
  y_pred = m*X + b
  plt.plot(X,y_pred)
plt.scatter(X,y)
plt.grid(True)
```
