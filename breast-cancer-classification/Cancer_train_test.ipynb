```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import numpy as np
```

```python
df = pd.read_excel("Cancer_train.xlsx")
x = df.drop(['id', 'diagnosis'], axis=1)  # # Features all measurement columns
y = df['diagnosis'].map({'M':1,'B':0})  # Label
```

```python
x.head()
```

```python
y.head()
```

```python
# Spliting the data into train-test
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size =0.2,random_state = 42)
```

```python
# Training the Random Forest
model = RandomForestClassifier(random_state = 42)
model.fit(x_train.values,y_train.values)
```

```python
# Prediction on test set
# Prediction only wants the features
prediction = model.predict(x_test.values)
prediction
```

```python
accuracy = accuracy_score(y_test,prediction)
print("Test set Accuracy : ",accuracy*100)
```

```python
#A confusion matrix is just a table that shows how many predictions were correct vs wrong for each class
cm = confusion_matrix(y_test,prediction)
cm
```

```python
#the classification report is basically an expanded version of the confusion matrix. It gives more detailed metrics for each class, not just overall accuracy. Let me explain clearly
print(classification_report(y_test,prediction))
```
